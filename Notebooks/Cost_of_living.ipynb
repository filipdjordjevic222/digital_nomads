# Cost of Living Cities


```python
regional_cost_df = pd.read_csv('movehubqualityoflife.csv')
```

```python
regional_cost_df.head()
```

```python
cities = [
   'Lisbon',
    'Tallinn',
    'Tbilisi',
    'Madrid',
    'New York City',
    'Bangkok',
    'Belgrade',  
    'Hanoi',
    'Manila',
    'Mexico City',
    'BogotÃ¡',
    'Montevideo',
    'Buenos Aires'
]
```

```python
regional_cost_clean_df = regional_cost_df[regional_cost_df['City'].isin(cities)]
regional_cost_clean_df = regional_cost_clean_df.reset_index(drop=True)
```

```python
regional_cost_clean_df
```

Movehub Rating: A combination of all scores for an overall rating for a city or country.
Purchase Power: This compares the average cost of living with the average local wage.
Health Care: Compiled from how citizens feel about their access to healthcare, and its quality.
Pollution: Low is good. A score of how polluted people find a city, includes air, water and noise pollution.
Quality of Life: A balance of healthcare, pollution, purchase power, crime rate to give an overall quality of life score.
Crime Rating: Low is good. The lower the score the safer people feel in this city.

```python
city_to_country = {
    'Mexico City': 'Mexico',
    'Manila': 'Philippines',
    'Buenos Aires': 'Argentina',
    'Hanoi': 'Vietnam',
    'Montevideo': 'Uruguay',
    'Lisbon': 'Portugal',
    'Madrid': 'Spain',
    'Belgrade': 'Serbia',
    'Bangkok': 'Thailand'
}
```

```python
regional_cost_clean_df['Country'] = regional_cost_clean_df['City'].map(city_to_country)
# Move the 'Country' column to the first position
regional_cost_clean_df = regional_cost_clean_df[['Country'] + [col for col in regional_cost_clean_df.columns if col != 'Country']]
```

```python
regional_cost_clean_df
```

```python
regional_cost_clean_df = regional_cost_clean_df.reset_index(drop=True)
regional_cost_clean_df = regional_cost_clean_df.sort_values('Country', ascending=True)
```

```python
regional_cost_clean_df
```

```python
regional_cost_clean_df.to_csv('cost_of_living_cities_clean.csv', index=False)
```

# Global income purchase power


```python
global_income_df = pd.read_csv('worldwide.csv')
global_income_df 
```

```python
global_income_df = global_income_df.rename(columns={'Country/Region': 'Country'})
```

```python
countries = [
    'Portugal', 'Estonia', 'Georgia', 'Spain', 'United States', 
    'Thailand', 'Serbia', 'Vietnam', 'Philippines', 'Mexico', 
    'Colombia', 'Uruguay', 'Argentina'
]
```

```python
global_income_clean_df = global_income_df[global_income_df['Country'].isin(countries)]
```

```python
global_income_clean_df = global_income_clean_df.sort_values('Country', ascending=True)
global_income_clean_df = global_income_clean_df.reset_index(drop=True)
global_income_clean_df = global_income_clean_df.drop(columns=['Rank'])
```

```python
global_income_clean_df
```

```python
global_income_clean_df.to_csv('global_income_clean.csv', index=False)
```
