```python
countries_and_capitals = {
    'Portugal': 'Lisbon',
    'Estonia': 'Tallinn',
    'Georgia': 'Tbilisi',
    'Spain': 'Madrid',
    'United States':'New York City',
    'Thailand': 'Bangkok',
    'Serbia': 'Belgrade',  
    'Vietnam': 'Hanoi',
    'Philippines': 'Manila',
    'Mexico': 'Mexico City',
    'Colombia': 'Bogot√°',
    'Uruguay': 'Montevideo',
    'Argentina': 'Buenos Aires'
}
```

```python
weather_countries_df = weather_df[weather_df['Country'].isin(countries_and_capitals.keys()) & 
                 weather_df['City'].isin(countries_and_capitals.values())]
```

```python
weather_countries_df.head()
```

```python
#weather_countries_df.dtypes
```

```python
#reset index and drop ref.column
weather_countries_df = weather_countries_df.drop(columns=['Ref.'])
weather_countries_df= weather_countries_df.reset_index(drop=True)
```

```python
#Just in Celsius temperature
def extract_celsius(temp_str):
   
    temp_str = temp_str.split('\n')[0]  # Get the part before the newline
    temp_str = temp_str.replace('‚àí', '-')  # Replace the special minus sign with the standard one
    temp_str = temp_str.strip()  # Remove any leading/trailing whitespace
    try:
        return float(temp_str)  # Convert to float
    except ValueError:
        return None 
```

```python
temperature_columns = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'Year']
for col in temperature_columns:
    weather_countries_df[col] = weather_countries_df[col].apply(extract_celsius)
```

```python
#weather_countries_df
```

```python
weather_countries_df = weather_countries_df.sort_values("Country", ascending=True)
weather_countries_df = weather_countries_df.reset_index(drop=True)
```
