# Crime Rate

```python
import pandas as pd

crime_rate_df = pd.read_csv('../../Data/Raw/crime-rate-by-country-2023.csv')
```

```python
new_crime_rate_df = crime_rate_df.rename(columns={'country': 'Country', 'crimeIndex': 'Crime Index', 'pop2023':'Population'})
```

```python
new_crime_rate_df
```

```python
countries = [
    'Portugal', 'Estonia', 'Georgia', 'Spain', 
    'Thailand', 'Serbia', 'Vietnam', 'Philippines', 'Mexico', 
    'Colombia', 'Uruguay', 'Argentina'
]
```

```python
crime_rate_clean_df = new_crime_rate_df[new_crime_rate_df['Country'].isin(countries)]
```

```python
crime_rate_clean_df= crime_rate_clean_df.sort_values('Country', ascending=True)
crime_rate_clean_df= crime_rate_clean_df.reset_index(drop=True)
```

```python
crime_rate_clean_df
```

```python
clean_crime_rate_df=crime_rate_clean_df.drop(columns=["rank"])
```
