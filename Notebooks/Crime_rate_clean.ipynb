# CrimeRate

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crime_df = pd.read_csv("crime_rate_clean-Copy1.csv")
crime_df 
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(x='Country', y='Crime Index', data=crime_df)
plt.title('Crime Rate by Country')
plt.xticks(rotation=45)
plt.xlabel('Country')
plt.ylabel('Crime Rate Index')
plt.tight_layout()

#save
plt.savefig('crime_rate_country.png', dpi=300)
plt.show()
```

# top 5 and lowest 5

```python

crime_sorted_df = crime_df.sort_values(by='Crime Index', ascending=True)
crime_sorted_df 
```

# top5

```python

top_5 = crime_sorted_df.head(5)
last_5 = crime_sorted_df.tail(5)

top_5
```

```python
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")

# 1. Bar Plot for Crime Index comparison
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
sns.barplot(x='Crime Index', y='Country', data=top_5)
plt.title('Lowest Crime Index Top 5 Countries')

plt.savefig('crime_index_top5.png', dpi=300)
plt.show()
```

```python
#population and crime rate index
plt.subplot(1, 2, 2)  
sns.scatterplot(x='Crime Index', y='Country', size='Population', sizes=(100, 1000), data=top_5, legend=None)
plt.title('Lowest Crime Index vs. Population')
plt.xlabel('Crime Index')
plt.ylabel('Country')

plt.savefig('crime_index_top5vsPopulation.png', dpi=300)
plt.show()
```

# Last5

```python
last_5 
```

```python
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")


plt.subplot(1, 2, 1) 
sns.barplot(x='Crime Index', y='Country', data=last_5)
plt.title('Highest Crime Index Top 5 Countries')

plt.savefig('highest_crime_index_top5.png', dpi=300)
plt.show()
```

```python
# Crime rate and population
plt.subplot(1, 2, 2)  
sns.scatterplot(x='Crime Index', y='Country', size='Population', sizes=(100, 300), data=last_5, legend=None)
plt.title('Hishest Crime Index vs. Population')
plt.xlabel('Crime Index')
plt.ylabel('Country')

plt.savefig('Highest_crime_index_top5vsPopulation.png', dpi=300)
plt.show()
```
